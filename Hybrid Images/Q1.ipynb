```python
import numpy
from PIL import Image
from numpy.fft import fft2, ifft2, fftshift, ifftshift
from scipy import misc
from scipy import ndimage
import matplotlib.pyplot as plt
import math
```

```python
def generateGaussianFilter(numRows, numCols, sigma, highPass=True):
   centerI = int(numRows/2) + 1 if numRows % 2 == 1 else int(numRows/2)
   centerJ = int(numCols/2) + 1 if numCols % 2 == 1 else int(numCols/2)

   def gaussian(i,j):
      coefficient = math.exp(-1.0 * ((i - centerI)**2 + (j - centerJ)**2) / (2 * sigma**2))
      return 1 - coefficient if highPass else coefficient

   return numpy.array([[gaussian(i,j) for j in range(numCols)] for i in range(numRows)])
```

```python
def applyFilter(image, filterMatrix):
   filteredImg = fftshift(fft2(image)) * filterMatrix
   filteredImg = ifft2(ifftshift(filteredImg))
   return filteredImg
```

```python
def takeHighFreq(image, sigma):
   return applyFilter(image, generateGaussianFilter(image.shape[0], image.shape[1], sigma, highPass=True))
```

```python
def takeLowFreq(image, sigma):
   return applyFilter(image, generateGaussianFilter(image.shape[0], image.shape[1], sigma, highPass=False))
```

```python
def outputPlot(img):
    plt.figure(1)
    plt.title('Hybrid_Image')
    plt.imshow(img)
    plt.show()
```

```python
high_freq_image = numpy.array(Image.open("HW1_Q1/bird.bmp"))
low_freq_image = numpy.array(Image.open("HW1_Q1/plane.bmp"))
alpha = 25
beta = 15
if(high_freq_image.shape[2]):
    r = high_freq_image[:,:,0]
    high_freq_image[:,:,0] = takeHighFreq(r,alpha)
    g = high_freq_image[:,:,1]
    high_freq_image[:,:,1] = takeHighFreq(g,alpha)
    b = high_freq_image[:,:,2]
    high_freq_image[:,:,2] = takeHighFreq(b,alpha)
    r = low_freq_image[:,:,0]
    low_freq_image[:,:,0] = takeLowFreq(r,beta)
    g = low_freq_image[:,:,1]
    low_freq_image[:,:,1] = takeLowFreq(g,beta)
    b = low_freq_image[:,:,2]
    low_freq_image[:,:,2] = takeLowFreq(b,beta)
    outputPlot(high_freq_image + low_freq_image)
else :
    high_freq_image = highPass(high_freq_image, alpha)
    low_freq_image = lowPass(low_freq_image, beta)
    outputPlot(high_freq_image + low_freq_image)
```
